```python
import pandas as pd
import datetime
import random
```

```python
df = pd.read_csv('2018_sa.csv')
```

```python
df
```

```python
df.drop(['id_str','user_full_name','date_column','lang','time'],inplace=True, axis=1)
df
```

```python
df.drop('preprocessed_text',inplace=True, axis=1)
df
```

```python
df.drop('user_verified',inplace=True, axis=1)
df
```

```python
df1 = pd.read_csv('2019_sa.csv')
df1
```

```python
# convert the 'Date' column to datetime format
df1['date'] = pd.to_datetime(df1['date'])

# change the date format to yyyy-mm-dd
df1['date'] = df1['date'].dt.strftime('%Y-%m-%d')
```

```python
df1
```

```python
df1.drop(['id_str','lang','user_full_name','time','preprocessed_text'],axis=1, inplace=True)
```

```python
df1
```

```python
df1 = df1.dropna(axis=1)
df1
```

```python
df2 = pd.read_csv('2020_sa.csv')
df2
```

```python
df2.drop(['id_str','lang','user_full_name','user_verified','time','preprocessed_text','date_column'],axis=1, inplace=True)
```

```python
df2
```

```python
# concatenate the two DataFrames
result = pd.concat([df,df1,df2])

result
```

```python
result.to_csv('tweet_sentiment_analysis.csv',index=False)
```
