```python
import pandas as pd

# Load the Tesla stock data into a pandas DataFrame
tesla_data = pd.read_csv('TSLA_2020_2021.csv', index_col='Date', parse_dates=True)

# Resample the data on business days
tesla_data = tesla_data.resample('B').last()

# Forward fill any missing data
tesla_data = tesla_data.ffill()
```
